# Problemas de distribución de probabilidad

## Distribución binomial

En teoría de la probabilidad y estadística, la distribución binomial es la distribución de probabilidad discreta que da sólo dos resultados posibles en un experimento; por ejemplo, al lanzar una moneda, puede salir cara o cruz.

```python
from scipy.stats import binom

# Definir todas las funciones de probabilidad relacionadas con esta distribución.

def dbinom(x, size, prob = 0.5):
    """
    Calcula la estimación puntual de la distribución binomial.
    """
    result = binom.pmf(k = x, n = size, p = prob, loc = 0)

    return result

def pbinom(q, size, prob = 0.5):
    """
    Calcula el acumulado de la distribución binomial.
    """
    result = binom.cdf(k = q, n = size, p = prob, loc = 0)

    return result

def qbinom(p, size, prob = 0.5):
    """
    Calcula la función cuantil a partir de la distribución binomial.
    """
    result = binom.ppf(q = p, n = size, p = prob, loc = 0)

    return result

def rbinom(n, size, prob = 0.5):
    """
    Genera variables aleatorias a partir de la distribución binomial.
    """
    result = binom.rvs(n = size, p = prob, size = n)

    return result
```

```python
# Distribución binomial(10, 0.2), dónde n = 10 es el número de ensayos y p = 0.2 es la probabilidad de éxito en un solo ensayo
import numpy as np

np.random.seed(42)

print(f"Probability that a Binomial(10, 0.2) takes the value 2: {dbinom(2, size = 10, prob = 0.2)}")
print(f"Probability that a Binomial(10, 0.2) will take a value less than 2: {pbinom(2, size = 10, prob = 0.2)}")
print(f"Which value of a Binomial(10, 0.2) has a cumulative probability of 0.9? {qbinom(0.9, size = 10, prob = 0.2)}")
print(f"Generate 2000 random values from a Binomial(10, 0.2) distribution: {rbinom(2000, size = 10, prob = 0.2)}")
```

```python
import matplotlib.pyplot as plt

np.random.seed(42)

size = 10
prob = 0.2
data = rbinom(2000, size = 10, prob = 0.2)

pmf = dbinom(data, size = size, prob = prob)
cdf = pbinom(data, size = size, prob = prob)
ppf = qbinom(data, size = size, prob = prob)

fig, axis = plt.subplots(3, 1, figsize = (5, 7))

axis[0].stem(data, pmf, "r-")
axis[1].stem(data, cdf, "b-")
axis[2].stem(data, ppf, "g-")

# Adjust the layout
plt.tight_layout()

# Show the plot
plt.show()
```

DISTRIBUCIÓN NORMAL

```python
from scipy.stats import norm

def dnorm(x, size = 0, prob = 1):
    """
    Calcula la densidad de probabilidad de la distribución normal.
    """
    result = norm.pdf(x = x, loc = size, scale = prob)

    return result


def pnorm(q, size = 0, prob = 1):
    """
    Calcula la probabilidad acumulada de la distribución normal.
    """
    result = norm.cdf(x = q, loc = size, scale = prob)

    return result


def qnorm(p, size = 0, prob = 1):
    """
    Calcula el cuantil de la distribución normal.
    """
    result = norm.ppf(q = p, loc = size, scale = prob)

    return result


def rnorm(n, size = 0, prob = 1):
    """
    Genera variables aleatorias a partir de la distribución normal.
    """
    result = norm.rvs(loc = size, scale = prob, size = n)

    return result

import numpy as np

np.random.seed(42)

print(f"La densida de probabilidad de la distribución normal es de: {dnorm(2, size = 10, prob = 0.2)}")
print(f"La probabilidad acumulada de la distribución normal es de: {pnorm(2, size = 10, prob = 0.2)}")
print(f"El cuantil es igual a: {qnorm(0.9, size = 10, prob = 0.2)}")
print(f"Las variables aleatorias de la distribución normal es igual a: {rnorm(2000, size = 10, prob = 0.2)}")
```

```python
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(42)

size = 0
prob = 1
data = rnorm(2000, size = size, prob = prob)

pdf = dnorm(data, size = size, prob = prob)
cdf = pnorm(data, size = size, prob = prob)
pmf = qnorm(cdf, size = size, prob = prob)

fig, axis = plt.subplots(3, 1, figsize = (5, 7))

axis[0].plot(data, pdf, "r-")
axis[1].plot(data, cdf, "b-")
axis[2].plot(data, pmf, "g-")

plt.tight_layout()

plt.show()
```

DISTRIBUCIÓN UNIFORME

```python
from scipy.stats import uniform

def dunif(x, min = 0, max = 1):
    """
    Calcula la densidad de probabilidad de la distribución uniforme.
    """
    result = uniform.pdf(x = x, loc = min, scale = max - min)

    return result


def punif(q, min = 0, max = 1):
    """
    Calcula el acumulado de la distribución uniforme.
    """
    result = uniform.cdf(x = q, loc = min, scale = max - min)

    return result


def qunif(p, min = 0, max = 1):
    """
    Calcula la función cuantil de la distribución uniforme.
    """
    result = uniform.ppf(q = p, loc = min, scale = max - min)

    return result


def runif(n, min = 0, max = 1):
    """
    Genera variables aleatorias a partir de la distribución uniforme.
    """
    result = uniform.rvs(loc = min, scale = max - min, size = n)

    return result

import numpy as np

np.random.seed(42)

print(f"La densidad de probabilidad de la distribución uniforme es: {dunif(0.5, min = 0, max = 1)}")
print(f"La probabilidad acumulada de la distribución uniforme es: {punif(0.5, min = 0, max = 1)}")
print(f"El cuantil de la distribución uniforme es: {qunif(0.9, min = 0, max = 1)}")
print(f"Las variables aleatorias generadas de la distribución uniforme son: {runif(10, min = 0, max = 1)}")
```

```python
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(42)

min = 0
max = 1
data = runif(2000, min = min, max = max)

pdf = dunif(data, min = min, max = max)
cdf = punif(data, min = min, max = max)
pmf = qunif(cdf, min = min, max = max)

fig, axis = plt.subplots(3, 1, figsize = (5, 7))

axis[0].plot(data, pdf, "r-")
axis[1].plot(data, cdf, "b-")
axis[2].plot(data, pmf, "g-")

plt.tight_layout()

plt.show()
```
